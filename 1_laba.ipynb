# Безынерционное звено

```python
def shit(w,s):
    SIZE=(10,10)
    plt.figure(1,figsize=SIZE)
    y,x=step(w)
    z,c=impulse(w)
    plt.subplot(2,1,1)
    plt.title('Переходная функция') 
    plt.ylabel('Амплитуда')
    plt.xlabel('Время(с)')
    plt.grid(True)
    plt.plot(x,y,"b")
    y,x=step(s)
    plt.plot(x,y,'g')
    plt.subplot(2,1,2)
    plt.title('Импульсная функция') 
    plt.ylabel('Амплитуда')
    plt.xlabel('Время(с)')
    plt.grid(True)
    plt.plot(z,c,"b")
    z,c=impulse(s)
    plt.plot(z,c,"g")

    plt.figure(figsize=SIZE)
    bode(w,color='r')
    bode(s,color='g')
    
    plt.show()
```

```python
from control.matlab import *
import matplotlib.pyplot as plt
k=3
w=tf([k],[1e-10,1])
s=tf([k/2],[1e-10,1])
print(w)

shit(w,s)
```
